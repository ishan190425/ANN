```python
#import libaries
import pandas as pd
import numpy as np4
import tensorflow as tf
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
le = LabelEncoder()
sc = StandardScaler()
```

```python
#import dataset
data = pd.read_csv("Churn_Modelling.csv")

x = data.iloc[:,3:-1].values
y = data.iloc[:, -1].values

x[:,2] = le.fit_transform(x[:,2]) #simple change female = 0 male = 1
```

```python
#need to hot encode since France - 0 Spain - 1 Doesnt meant Spain is better than France
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(),[1])],remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
x
```

```python
#Split into training set and Test Set
X_train, X_test, y_train,y_test = train_test_split(x,y,test_size = .2, random_state = 0)
```

```python
#Feature Scaling (NEED WHENEVER WE USE DEEP LEARNING)
X_train = sc.fit_transform(X_train) #applies feature scaling
X_test = sc.fit_transform(X_test)
```

```python
#Part 2
```

```python
#Building the ANN
#Use a sequence of Layers (vs Computational Learning)
ann = tf.keras.models.Sequential()
```

```python
#add the input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation ='relu')) #input is the units (Number of nuerons in the first Layer) 
#(input layer is number of parameters) + hidden layers
#activation functions
#relu - rectifier function
```

```python
ann.add(tf.keras.layers.Dense(units = 6, activation ='relu')) #second layer, can change nuerons 
```

```python
ann.add(tf.keras.layers.Dense(units = 1, activation ='sigmoid')) #output layer, plus prob of 1, softmax when classifi
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) #using stoachstic Gradient Descent, #loss must be binary 
#when its a binary func
#categorical_crossentropy
```

```python
ann.fit(X_train,y_train,batch_size = 32, epochs = 1000)
```

```python
ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]])) #must use the same scaling, predict a .025 chance
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
